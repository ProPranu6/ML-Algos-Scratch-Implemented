```python
class node:
    def __init__(self,val):
        self.val=val
        self.right=None
        self.left=None
    




class binary_tree:
    
    def __init__(self,root):
        self.root=node(root)
        
    
    
    def pre(self,start):
        if start == None:
            return ""
            
        else:
            
            return  str(start.val) + "--" + self.pre(start.left) + self.pre(start.right)
    def post(self,start):
        if start == None:
            return ""
            
        else:
            
            return   self.post(start.left) + self.post(start.right) + str(start.val) + "--" 
            
    def ino(self,start):
        if start == None:
            return ""
            
        else:
            
            return   self.ino(start.left) + str(start.val) + "--" + self.ino(start.right) 
            
    def show(self,typ):
        if typ=="preorder":
            return self.pre(tree.root)
        elif typ=="postorder":
            return self.post(tree.root)
        elif typ=="inorder":
            return self.ino(tree.root)
        else:
            return "NO SUCH TRAVERSAL"
        
    
            


tree=binary_tree(1)
tree.root=node(30)
tree.root.left=node(20)
tree.root.right=node(39)
tree.root.left.left=node(10)
tree.root.left.left.right=node(15)
tree.root.left.right=node(25)
tree.root.left.right.left=node(23)
tree.root.right.left=node(35)
tree.root.right.right=node(42)

tree.show("postorder")

#in order:     10--15--20--23--25--30--35--39--42--
#pre order:    30--20--10--15--25--23--39--35--42--
#post order:   15--10--23--25--20--35--42--39--30--




        
```

```python
#IMPURITY CHECKING FUNCTIONS -----------------------------------START HERE--------------------------------------------------------------

def pimp(prcolumn):
    cs=0
    cno=0

    for i in re[prcolumn]:
        if i == 1 :
            cs+=1
        else:
            cno+=1

    cshs=0
    cshno=0
    cnohs=0
    cnohno=0

   
    for (i,j) in enumerate(re[prcolumn]):
        
        if j == 1 and re[re.columns[16]][i]==1:
            cshs+=1
        elif j == 1 and re[re.columns[16]][i]==1 :
            cshno+=1
        elif j == 1 and re[re.columns[16]][i]==1 :
            cnohs+=1
        elif j == 0 and re[re.columns[16]][i]==0 :
            cnohno+=1
    cimp= (cs/(cs+cno))*(1 - (cshs/cs)**2 -(cshno/cs)**2) + (cno/(cs+cno))*(1 - (cnohno/cno)**2 -(cnohs/cno)**2)
    leftimp=(1 - (cshs/cs)**2 -(cshno/cs)**2)
    rightimp=(1 - (cnohno/cno)**2 -(cnohs/cno)**2)
    di={'timp':cimp,'leftimp':leftimp,'rightimp':rightimp,'ptT':cshs/cs,'pfT':cnohs/cno}
    return di




def simp(pacolumn,prcolumn):
    ptars=0
    ptarno=0

    for (i,j) in enumerate(re[prcolumn]):
        if j == 1 and re[pacolumn][i]==1 :
            ptars+=1
        elif j == 0 and re[pacolumn][i]==1:
            ptarno+=1
     
    ptarshs=0
    ptarshno=0
    ptarnohs=0
    ptarnohno=0
    for (i,j) in enumerate(re[pacolumn]):
        
        if j == 1 and re[re.columns[16]][i]==1 and re[prcolumn][i]==1 :
            ptarshs+=1
        elif j == 1 and re[re.columns[16]][i]==0 and re[prcolumn][i]==1:
            ptarshno+=1
        elif j == 1 and re[re.columns[16]][i]==1 and re[prcolumn][i]==0:
            ptarnohs+=1
        elif j == 1 and re[re.columns[16]][i]==0 and re[prcolumn][i]==0 :
            ptarnohno+=1
    ptarimp= (ptars/(ptars+ptarno))*(1 - (ptarshs/ptars)**2 -(ptarshno/ptars)**2) + (ptarno/(ptars+ptarno))*(1 - (ptarnohno/ptarno)**2 -(ptarnohs/ptarno)**2)
    ptleftimp=(1 - (ptarshs/ptars)**2 -(ptarshno/ptars)**2)
    ptrightimp=(1 - (ptarnohno/ptarno)**2 -(ptarnohs/ptarno)**2)
    
    pfars=0
    pfarno=0

    for (i,j) in enumerate(re[prcolumn]):
        if j == 1 and re[pacolumn][i]==0:
            pfars+=1
        elif j == 0 and re[pacolumn][i]==0:
            pfarno+=1
        
    pfarshs=0
    pfarshno=0
    pfarnohs=0
    pfarnohno=0
    for (i,j) in enumerate(re[pacolumn]):
        if j == 0 and re[re.columns[16]][i]==1 and re[prcolumn][i]==1 :
            pfarshs+=1
        elif j == 0 and re[re.columns[16]][i]==0 and re[prcolumn][i]==1:
            pfarshno+=1
        elif j == 0 and re[re.columns[16]][i]==1 and re[prcolumn][i]==0:
            pfarnohs+=1
        elif j == 0 and re[re.columns[16]][i]==0 and re[prcolumn][i]==0 :
            pfarnohno+=1
    pfarimp= (pfars/(pfars+pfarno))*(1 - (pfarshs/pfars)**2 -(pfarshno/pfars)**2) + (pfarno/(pfars+pfarno))*(1 - (pfarnohno/pfarno)**2 -(pfarnohs/pfarno)**2)
    pfleftimp=(1 - (pfarshs/pfars)**2 -(pfarshno/pfars)**2)
    pfrightimp=(1 - (pfarnohno/pfarno)**2 -(pfarnohs/pfarno)**2)
    
    di={'pttimp':ptarimp,'ptleftimp':ptleftimp,'ptrightimp':ptrightimp,'pttT':ptarshs/ptars,'ptfT':ptarnohs/ptarno,'pftimp':pfarimp,'pfleftimp':pfleftimp,'pfrightimp':pfrightimp,'pftT':pfarshs/pfars,'pffT':pfarnohs/pfarno}
    return di


#IMPURITY CHECKING FUNCTIONS------------------------------------------------END HERE-------------------------------------------


#CONDITIONS TO CONSIDER FOR A HEIGHT 2 BINARY TREE ----------------------------------------------------------------------------

if pimp(re.columns[15])['timp']<=pimp(re.columns[7])['timp']:
    
    tree=bt(re.columns[15])
    if simp(re.columns[15],re.columns[7])['pttimp']<= pimp(re.columns[15])['leftimp']:
        
        tree.root.lp=node(re.columns[7])
        if simp(re.columns[15],re.columns[7])['pttT']>=simp(re.columns[15],re.columns[7])['ptfT']:
            resl="Emergency"
            resr="Completely safe"
        else:
            resl="Completely safe"
            resr="Emergency"
        tree.root.lp.lp=node(resl)
        tree.root.lp.rp=node(resr)
    
    else:
        
        if pimp(re.columns[15])['ptT']>=pimp(re.columns[15])['pfT']:
            resl="Emergency"
            resr="Completely safe"
        else:
            resl="Completely safe"
            resr="Emergency"
        tree.root.lp=node(resl)
    
    if simp(re.columns[15],re.columns[7])['pftimp']<= pimp(re.columns[15])['rightimp']:
        
        tree.root.rp=node(re.columns[7])
        if simp(re.columns[15],re.columns[7])['pftT']>=simp(re.columns[15],re.columns[7])['pffT']:
            resl="Emergency"
            resr="Completely safe"
        else:
            resl="Completely safe"
            resr="Emergency"
        tree.root.rp.lp=node(resl)
        tree.root.rp.rp=node(resr)
    
    else:
        
        if pimp(re.columns[15])['ptT']>=pimp(re.columns[15])['pfT']:
            resl="Emergency"
            resr="Completely safe"
        else:
            resl="Completely safe"
            resr="Emergency"
        tree.root.rp=node(resr)
else:
    
    tree=bt(re.columns[7])
    if simp(re.columns[7],re.columns[15])['pttimp']<= pimp(re.columns[15])['leftimp']:
        
        tree.root.lp=node(re.columns[15])
        if simp(re.columns[7],re.columns[15])['pttT']>=simp(re.columns[7],re.columns[15])['ptfT']:
            resl="Emergency"
            resr="Completely safe"
        else:
            resl="Completely safe"
            resr="Emergency"
        tree.root.lp.lp=node(resl)
        tree.root.lp.rp=node(resr)
    
    else:
        
        if pimp(re.columns[7])['ptT']>=pimp(re.columns[7])['pfT']:
            resl="Emergency"
            resr="Completely safe"
        else:
            resl="Completely safe"
            resr="Emergency"
        tree.root.lp=node(resl)
    
    if simp(re.columns[7],re.columns[15])['pftimp']<= pimp(re.columns[7])['rightimp']:
        
        tree.root.rp=node(re.columns[15])
        if simp(re.columns[7],re.columns[15])['pftT']>=simp(re.columns[7],re.columns[15])['pffT']:
            resl="Emergency"
            resr="Completely safe"
        else:
            resl="Completely safe"
            resr="Emergency"
        tree.root.rp.lp=node(resl)
        tree.root.rp.rp=node(resr)
    
    else:
        
        if pimp(re.columns[7])['ptT']>=pimp(re.columns[7])['pfT']:
            resl="Emergency"
            resr="Completely safe"
        else:
            resl="Completely safe"
            resr="Emergency"
        tree.root.rp=node(resr)

#------------------------------------------------------------------------------------------------------------------------------

#USER INTERACTION FUNCTION ----------------------------------------------------------------------------------------------------

def ask(e,l):
    
    if e.rp== None or e.lp == None:
        
        return e.val
    
    else:
        
        res1=l[0]
        if res1== 1:
            return ask(e.lp,l[1:])
        
        elif res1==0:
            return ask(e.rp,l[1:])
        
        else:
            return "SOMETHING HAS GONE WRONG CHECK WHETHER IT'S YOUR TYPO"
#------------------------------------------------------------------------------------------------------------------------------

#PLEASE HAVE THIS KEY WORD WITH YOU AND PASS IT INTO THE ASK FUNCTION:
e=tree.root



print(ask(e,[0,0,1]))

```
